```python
# Load Processing libraries
import numpy as np
import pandas as pd

# Load Dataset
from sklearn import datasets

# Load Graphing library and presets
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
# Calling Dataset
data = datasets.load_wine()
```

```python
# Creating dataframe
df=pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
# First 5 rows in the dataset
df.head()
```

```python
df['Class']=data['target']
df['Class_target']=df['Class'].apply(lambda x: data['target_names'][x])
```

```python
df.head()
```

```python
# Statistical values describing the data
df.describe()
```

```python
# Data Columns datatypes
df.info()
```

```python
# Check for null values in the dataset
df.isnull().sum()
```

### **1. Plotting a Heatmap to show correlation between values**

```python

plt.figure(figsize=(15,15))

cor= df.corr()
sns.heatmap(cor, annot = True)
```

**Dropping loosely related values to the target value**

```python
df.drop(columns=['color_intensity','magnesium','ash','malic_acid','alcohol'],axis=1, inplace=True)
```

```python
df.head()
```

### **2. Plotting pair plot showing trends with all possible comparision**

```python
sns.pairplot(df)
```

**Let's Just look into highly correlated properties**
1. flavanoids
2. total_phenols
3. d280/od315_of_diluted_wines
4. hue
5. proline

### **3. Plotting Class values**

```python
sns.barplot(x="Class_target",y="total_phenols",data=df)
```

```python
sns.barplot(x="Class_target",y="flavanoids",data=df)
```

```python
sns.barplot(x="Class_target",y="proline",data=df)
```

#### Insights Found:
- Class 0 has the highest values of all properties  
- Class 1 comes second most of cases  
- Class 2 have the lowest values of all

### **4. Confirming insights found using scatter plot to see distribution**

```python
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
sns.scatterplot(x='total_phenols',y="flavanoids",hue="Class_target", data=df)

plt.subplot(2,2,2)
sns.scatterplot(x='total_phenols',y="od280/od315_of_diluted_wines",hue="Class_target", data=df)

plt.subplot(2,2,3)
sns.scatterplot(x='total_phenols',y="hue",hue="Class_target", data=df)

plt.subplot(2,2,4)
sns.scatterplot(x='total_phenols',y="proline",hue="Class_target", data=df)
```

#### Confirmations
1. Total Phenols and Flavanoids have a positive linear relationship.
2. Class 0 has the highest values of all classes.
3. Only Total phenols and flavanoids show a positive linear relationship. 
4. Class 0 is highly distinguisable using the higher values of total phenols and proline

### **5. Plotting Violin plot to see variance in dataset**

```python
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
sns.violinplot(x="Class_target",y="total_phenols" ,data=df)

plt.subplot(2,2,2)
sns.violinplot(x="Class_target",y="flavanoids" ,data=df)

plt.subplot(2,2,3)
sns.violinplot(x="Class_target",y="hue" ,data=df)

plt.subplot(2,2,4)
sns.violinplot(x="Class_target",y="proline" ,data=df)
```

#### Knowledge from violin plots:
1. Class 1 and Class 2 have highly similar values in case of proline.

### **6. Check for Outliers using box plot:**

```python
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
sns.boxplot(x="Class_target",y="total_phenols" ,data=df)

plt.subplot(2,2,2)
sns.boxplot(x="Class_target",y="flavanoids" ,data=df)

plt.subplot(2,2,3)
sns.boxplot(x="Class_target",y="hue" ,data=df)

plt.subplot(2,2,4)
sns.boxplot(x="Class_target",y="proline" ,data=df)
```

#### Outliers in Clases:
1. Class 0 has few outliers in case of total phenols.
2. Class 2 has outliers in case of total phenols and flavanoids.
3. Class 1 has outliers in three different cases flavanoids, hue and proline but, perfectly fits in case of total phenols
